```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
avocados = pd.read_csv('./dataSets/avocado.csv')
```

```python
avocados.head()  # Display the first few rows of the DataFrame
```

```python
avocados = avocados.drop(columns=['Unnamed: 0'])
```

```python
avocados = avocados.reset_index(drop=True)  # Reset the index of the DataFrame
```

```python
avocados.head()  # Display the first few rows of the DataFrame
```

```python
# obtener el total de numeros de aguacates vendidos por cada tamaño
nb_sold_by_type = avocados.groupby('type')[['Small Bags', 'Large Bags', 'XLarge Bags']].sum()
```

```python
nb_sold_by_type
```

```python
# crar una grafica de barras del numero de aguacates vendidos por tamaño
nb_sold_by_type.plot(kind='bar')
# Mostrar la grafica
plt.show()
```
