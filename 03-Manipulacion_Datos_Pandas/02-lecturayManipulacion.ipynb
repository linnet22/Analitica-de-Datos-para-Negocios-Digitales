# Filtrado con Operadores Logicos

```python
df[df['onsite_remote'] == 'remote']  # Filtra las filas donde la columna 'onsite_remote' es igual a 'remote'
```

```python
df[~(df['onsite_remote'] == 'remote')]  # Filtra las filas donde la columna 'onsite_remote' no es igual a 'remote'
```

# Tabajos que no sean de Amazon ni remotos

```python
df[~((df['company'] == 'Amazon ') & (df['onsite_remote'] == 'remote'))]
```
